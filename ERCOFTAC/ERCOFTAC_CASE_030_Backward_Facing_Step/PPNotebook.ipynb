```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
## cp
cp_exp = pd.read_csv('cp.csv')
cp_ph = pd.read_csv('lower_wall_ph.csv')
cp_star = pd.read_csv('lower_wall_star_p.csv')

H = 0.0127
cp_factor = 0.5*1.18415*48.18**2 # 1/2 rho U_ref^2 
# Create subplots

fig, ax1 = plt.subplots(1, 1, figsize=(4, 4))

cp_star = cp_star.sort_values(by='Region: Body 1.wall: Centroid[X] (m)')

ax1.plot(cp_exp['x/h'], cp_exp['cp'], 'k.', label='Exp')
ax1.plot(cp_ph['Points_0']/H, cp_ph['pressure']/cp_factor +0.125, 'b', label='MOOSE-NS')
ax1.plot(cp_star['Region: Body 1.wall: Centroid[X] (m)']/H, cp_star['Region: Body 1.wall: Pressure (Pa)']/cp_factor +0.125, 'r', label='Star-CCM')

ax1.set_xlabel(r'$\mathrm{x/h}$', fontsize=14)
ax1.set_ylabel(r'$\mathrm{c_p}$', fontsize=14)
ax1.set_xlim([-5,22])
ax1.legend()
ax1.grid(True)

# Adjust layout
plt.title('Pressure coefficient')
plt.tight_layout()

plt.savefig('plots_cp.pdf')
```

```python
##cf
cp_exp = pd.read_csv('cf.csv')
cp_ph = pd.read_csv('lower_wall_ph.csv')
cp_star = pd.read_csv('lower_wall_star.csv')
cp_star = cp_star.sort_values(by='Region: Body 1.wall: Centroid[X] (m)')

H = 0.0127
cp_factor = 0.5*1.18415*47.**2
# Create subplots

fig, ax1 = plt.subplots(1, 1, figsize=(4, 4))

ax1.plot(cp_exp['x/h'], cp_exp['cf'], 'k.', label='Exp')
ax1.plot(cp_ph['Points_0']/H, cp_ph['tau_w']/cp_factor, 'b', label='MOOSE-NS')
ax1.plot(cp_star['Region: Body 1.wall: Centroid[X] (m)']/H, cp_star['Region: Body 1.wall: Wall Shear Stress[i] (Pa)']/cp_factor, 'r', label='Star-CCM')

ax1.set_xlabel(r'$\mathrm{x/h}$', fontsize=14)
ax1.set_ylabel(r'$\mathrm{c_f}$', fontsize=14)
ax1.set_xlim([-5,22])
ax1.legend()
ax1.grid(True)

# Adjust layout
plt.title('Skin Friction coefficient')
plt.tight_layout()

plt.savefig('plots_cf.pdf')
```

```python
#u profiles
u_exp = pd.read_csv('u_profiles_exp.csv')
u_ph_1 = pd.read_csv('u+1_ph.csv')
u_ph_4 = pd.read_csv('u+4_ph.csv')
u_ph_6 = pd.read_csv('u+6_ph.csv')
u_ph_10 = pd.read_csv('u+10_ph.csv')
u_star_1 = pd.read_csv('u+1_star.csv')
u_star_1 = u_star_1.sort_values(by='Line Probe: Centroid[Y] (m)')
u_star_4 = pd.read_csv('u+4_star.csv')
u_star_4 = u_star_4.sort_values(by='Line Probe: Centroid[Y] (m)')
u_star_6 = pd.read_csv('u+6_star.csv')
u_star_6 = u_star_6.sort_values(by='Line Probe: Centroid[Y] (m)')
u_star_10 = pd.read_csv('u+10_star.csv')
u_star_10 = u_star_10.sort_values(by='Line Probe: Centroid[Y] (m)')
# Create subplots

fig, ax1 = plt.subplots(1, 1, figsize=(4, 4))

ax1.plot(u_exp['x+1'], u_exp['y/h'], 'kv', label='Exp x/h:1')
ax1.plot(u_ph_1['vel_0']/48.2, 1 + u_ph_1['Points_1']/0.0127, 'k', label='MOOSE-NS x/h:1')
ax1.plot(u_star_1['Line Probe: Velocity[i] (m/s)']/48.2, 1 + u_star_1['Line Probe: Centroid[Y] (m)']/0.0127, 'k--',markersize=2, label='Star-CCM x/h:1')

ax1.plot(u_exp['x+4'], u_exp['y/h'], 'rv', label='Exp x/h:4')
ax1.plot(u_ph_4['vel_0']/48.2, 1 + u_ph_4['Points_1']/0.0127, 'r', label='MOOSE-NS x/h:4')
ax1.plot(u_star_4['Line Probe: Velocity[i] (m/s)']/48.2, 1 + u_star_4['Line Probe: Centroid[Y] (m)']/0.0127, 'r--',markersize=2, label='Star-CCM x/h:4')

ax1.plot(u_exp['x+6'], u_exp['y/h'], 'bv', label='Exp x/h:6')
ax1.plot(u_ph_6['vel_0']/48.2, 1 + u_ph_6['Points_1']/0.0127, 'b', label='MOOSE-NS x/h:6')
ax1.plot(u_star_6['Line Probe: Velocity[i] (m/s)']/48.2, 1 + u_star_6['Line Probe: Centroid[Y] (m)']/0.0127, 'b--',markersize=2, label='Star-CCM x/h:6')

ax1.plot(u_exp['x+10'], u_exp['y/h'], 'gv', label='Exp x/h:10')
ax1.plot(u_ph_10['vel_0']/48.2, 1 + u_ph_10['Points_1']/0.0127, 'g', label='MOOSE-NS x/h:10')
ax1.plot(u_star_10['Line Probe: Velocity[i] (m/s)']/48.2, 1 + u_star_10['Line Probe: Centroid[Y] (m)']/0.0127, 'g--',markersize=2, label='Star-CCM x/h:10')

ax1.set_xlabel(r'$\mathrm{U/U_{ref}}$', fontsize=14)
ax1.set_ylabel(r'$\mathrm{y/h}$', fontsize=14)
ax1.set_ylim([-0.1,4.5])
ax1.legend(fontsize="6")
ax1.grid(True)

# Adjust layout
plt.title('Vertical x-Velocity Profiles')
plt.tight_layout()

plt.savefig('plots_u_profiles.pdf')
```
